```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

df = pd.read_csv('honeyproduction.csv')

plt.boxplot(df.prodvalue,showmeans=True)
plt.show()

plt.boxplot(df.priceperlb,showmeans=True)
plt.show()

plt.boxplot(df.yieldpercol,showmeans=True)
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=[10,10])
sns.boxplot(x='year',y='yieldpercol',data = df)
plt.show()
```

```python
i=86
y=[]
x=[]
while df.year[i] != '2000':
    y.append(df.prodvalue.loc[i])
    x.append(df.state.loc[i])
    i=i+1
    if i == 129:
        break
xs = [i for i, _ in enumerate(x)]
plt.figure(figsize=[10,10])
plt.bar(xs,y)
plt.xticks([i + 0.5 for i, _ in enumerate(x)], x,rotation = 90)
plt.xlabel("States in USA")
plt.ylabel("Production value of Honey")
plt.show()
```

```python
import seaborn as sns
sns.jointplot(x='priceperlb',y='yieldpercol',data = df,kind = 'reg')
plt.xlabel("Price per pound")
plt.ylabel("Yield per colony")
plt.show()
```
